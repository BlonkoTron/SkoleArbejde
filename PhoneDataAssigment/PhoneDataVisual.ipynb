```python
pData = pd.read_xml('PhoneData.xml')
print(pData)
```

```python
pData.describe(include = 'all')
```

```python
plt.plot(pData.x,pData.y)
plt.legend()
plt.grid(True)
plt.title('Movement of the phone on the x and y axis')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(pData.x,pData.z)
plt.legend()
plt.grid(True)
plt.title('Movement of the phone on the x and Z axis')
plt.xlabel('X')
plt.ylabel('Z')
```

```python
plt.plot(pData.z,pData.y)
plt.legend()
plt.grid(True)
plt.title('Movement of the phone on the Z and Y axis')
plt.xlabel('Z')
plt.ylabel('Y')
```

```python
plt.plot(pData.time,pData.x, label='X')
plt.plot(pData.time,pData.y, label='Y')
plt.plot(pData.time,pData.z, label='Z')

plt.legend()

plt.grid(True)

plt.title('Movement of the phone on each axis over time')

plt.xlabel('Time')

plt.ylabel('Value of each axis')
```
